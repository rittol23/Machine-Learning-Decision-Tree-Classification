```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
```

```python
TRDiseaseDataPath ='/content/Training.csv'
TestDataPath = '/content/Testing.csv'
```

```python
# read and store the dataframe

TrainData = pd.read_csv(TRDiseaseDataPath)
TestData = pd.read_csv(TestDataPath)
```

Select the predicted target and stick to convention by naming as 'y'

```python
y = TrainData.prognosis
```

Find the most active columns to see what is best as the feature list

```python
#drop our target
TrainDataTD = TrainData.drop(columns=['prognosis'])



# calculate the sum of each of the columns
TrainColSum = TrainDataTD.sum()

# put in descending order
SortedTrain = TrainColSum.sort_values(ascending=False)

# select the columns with the top 10 sums
MAFeatures = SortedTrain.head(10)
print(MAFeatures)

```

```python
MostActiveFeatures = ['fatigue', 'vomiting', 'high_fever', 'loss_of_appetite', 'nausea', 'headache', 'abdominal_pain', 'yellowish_skin', 'yellowing_of_eyes', 'chills']

X = TrainData[MostActiveFeatures]

X.describe()
```

```python
X.head()
```

```python
y.head()
```

**Build Model**

```python
# encode the prognosis labels with a unique integer we can decode later
from sklearn.preprocessing import LabelEncoder

#create label encoder object
label_encoder = LabelEncoder()

#fit the encoder
label_encoder.fit(y)

# transform the column to integer values using the fitted label encoder
y_encoded = label_encoder.transform(y)

print(y_encoded)
print(y)

```

```python

DiseasePredModel = tree.DecisionTreeClassifier(random_state=1)
DiseasePredModel.fit(X,y_encoded)
```

```python
print("Making predictions for the following symptoms:")
print(X.head())
print("The predicted diseases are:")

b4con = DiseasePredModel.predict(X.head())

string_y = label_encoder.inverse_transform(b4con)

print(string_y)
print(b4con)
```
